# Part One: Decomposition

## Chosen Research Area:  
# Product Analysis

## Research Objective:  
# The goal of this project is to analyze product performance for AtliQ Hardware across different regions and time periods. By identifying bestselling products, uncovering seasonal trends, and highlighting underperforming categories, this research aims to deliver data-driven insights that can support strategic decisions to improve sales and optimize product offerings.

---

## Potential Research Questions:

# 1. Which products generate the highest total revenue overall, and in which regions?
# 2. Which products are the most profitable and consistently high-performing?
# 3. What seasonal or monthly patterns can be observed in product sales?
# 4. What are the top underperforming products, and what factors might explain their lack of success?

---

## Hypotheses:

### 1. By Category
# Null Hypothesis: All product categories have the same average monthly revenue.
# Alternative Hypothesis: At least one product category has significantly different average monthly revenue.

### 2. New vs Old Products
# H0: There is no difference in average revenue between newly launched and older products.
# H1: Newly launched products generate significantly more revenue than older products.

### 3. By Region
# H0: Product revenue does not differ significantly across sales regions.
# H1: Product revenue significantly varies between sales regions.

# Total revenue by product
